```python
import glob
import re
import json
import ntpath

from flair.data import Sentence
from flair.models import SequenceTagger
import spacy
from spacy.tokens import Doc
import geocoder

from typing import List, Dict

from correct_ocr import single_characters, delete_specials, correct_s


import nltk
```

```python
# Load NER model into flair tagger.
tagger: SequenceTagger = SequenceTagger.load("dbmdz/flair-historic-ner-onb")
```

```python

nlp = spacy.load("de_core_news_md")
```

```python
import os
doc: Doc = nlp(open('../data/18th_century_first_quarter_corr/Z69804407.txt', 'r').read())
print(os.path.basename(re.sub('.txt', '.json', '../data/18th_century_first_quarter_corr/Z69804407.txt')))
```

```python
sent_tokens = nltk.sent_tokenize(open('../data/18th_century_first_quarter_corr/Z69804407.txt', 'r').read(), language='german')
print(len(sent_tokens))

sentence = [Sentence(i) for i in sent_tokens]
tagger.predict(sentence)

for sent in sentence:
    print(sent.to_tagged_string())
```
